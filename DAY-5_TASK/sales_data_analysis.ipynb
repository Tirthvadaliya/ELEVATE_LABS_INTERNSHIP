```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load CSV using Pandas
df = pd.read_csv('sales_data.csv')
```

```python
# 2. Basic Data Overview
print("First 5 rows:\n", df.head())
print("\nSummary statistics:\n", df.describe())
print("\nMissing values:\n", df.isnull().sum())

```

```python
# 3. Group by category and sum revenue
category_sales = df.groupby('category')['revenue'].sum().reset_index()
print("\nTotal Revenue by Category:\n", category_sales)

```

```python
# 4. Group by product and sum quantity sold
product_sales = df.groupby('product')['quantity'].sum().reset_index()
print("\nTotal Quantity Sold by Product:\n", product_sales)

```

```python
# 5. Plot revenue by category
plt.figure(figsize=(8,6))
plt.bar(category_sales['category'], category_sales['revenue'], color='skyblue')
plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Total Revenue')
plt.tight_layout()
plt.show()

```

```python
# 6. Plot quantity sold by product
plt.figure(figsize=(8,4))
plt.bar(product_sales['product'], product_sales['quantity'], color='orange')
plt.title('Quantity Sold by Product')
plt.xlabel('Product')
plt.ylabel('Total Quantity')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
